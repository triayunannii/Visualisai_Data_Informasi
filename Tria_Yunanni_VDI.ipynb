```python
df = pd.read_csv("Data_Tanaman_Padi_Sumatera_version_1.csv")
df.head()
```

# 1. Tren Produksi Padi per Provinsi dan Tahun.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Tahun', y='Produksi', hue='Provinsi', marker='o')
plt.title('Tren Produksi Padi per Provinsi dari Tahun ke Tahun')
plt.ylabel('Produksi (Ton)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Fungsi dalam Visualisasi**
Fungsi visualisasi ini adalah untuk memberikan informasi apakah dengan melakukan perluasan lahan selalu menghasilkan peningkatan produksi.

# **Tone dalam Visual**
Menggunakan lineplot dengan warna yang berbeda untuk mewakili Provinsi serta luas lahan panen dan produksinya yang memiliki tampilan sederhana dan mudah di pahami



# 2. Hubungan antara Produksi dengan Luas Panen.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Luas Panen', y='Produksi', hue='Provinsi', size='Luas Panen', sizes=(20, 200))
plt.title('Hubungan antara Produksi dan Luas Panen')
plt.xlabel('Luas Panen (Hektar)')
plt.ylabel('Produksi (Ton)')
plt.tight_layout()
plt.show()
```

# **Fungsi dalam Visual**
Fungsi visual ini memberikan informasi apakah perluasan lahan selalu menghasilkan peningkatan produksi

# **Tone dalam Visual**
Menggunakan scatter plot dengan ukuran dan warna yang berbeda untuk mewakili provinsi serta luas lahan panen dan produksinya,dengan tampilan yang menarik.

# 3. Perbandingan Total Produksi antar Provinsi

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Provinsi', y='Produksi', estimator=sum, ci=None)
plt.title('Total Produksi Padi per Provinsi')
plt.ylabel('Produksi (Ton)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Fungsi dalam Visualisasi**
Fungsi dari visualisasi ini adalah untuk menunjukkan perbedaan yang signifikan dalam jumlah produksi antar provinsi, sehingga para pemilik kepentingan dapat mengidentifikasi provinsi mana yang memproduksi lebih banyak beras dan mana yang memproduksi lebih sedikit. Hal ini dapat membantu mengarahkan sumber daya ke daerah-daerah yang membutuhkan perbaikan.

Menggunakan barchart yang memberikan informasi bersifat komparatif dan sederhanakesan tentang perbedaan antar provinsi
